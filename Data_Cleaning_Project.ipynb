# Data Cleaning With Python Using Pandas ..

```python
import pandas as pd # Importing pandas library for data cleaning and transforming our dataset
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
df
```

## Removing duplicates and save it to df variable

```python
df.duplicated() # Check if there duplicated rows and we find one row duplicated
df = df.drop_duplicates() # removing duplicates and save it to df variable
df
```

## Remove not useful column from our dataset

```python
df = df.drop(columns = 'Not_Useful_Column') # remove not useful column from our dataset
df
```

## last name column need to be standardize so we using Strip function to remove unneeded symbols or characters

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/') # last name column need to be standardize so we using Strip function to remove unneeded symbols or characters
df
```

# Cleaning and standardizing Phone Number column


```python
# Cleaning and standardizing Phone Number column
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
# df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

## Splitting Address Column

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2,  expand=True)
df
```

```python
# Standardizing Paying Customer and Do Not Contact Columns
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
# df = df.replace('N/a', '')

df = df.fillna('')
df
```

```python
# Drop rown with "Y" on Do Not Contact column
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace=True)

df
```

```python
# Drop rows with no phone number on it
for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace=True)

df
```

```python
df = df.reset_index(drop=True) # Reset index number to reorganize our dataset
df
```

```python
# Now we can remove Address column cause we already split it to three columns
df = df.drop(columns = 'Address')
df
```
